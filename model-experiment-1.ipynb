# removing not important features, these features are selected by me, not other basis.

```python
important_features = [
    'OverallQual', 'GrLivArea', 'TotalBsmtSF', 'GarageCars', 'YearBuilt',
    'Neighborhood', 'FullBath', 'KitchenQual', 'ExterQual', 'BsmtQual',
    'LotArea', 'YearRemodAdd', 'MSZoning', 'FireplaceQu', 'Fireplaces',
    'BedroomAbvGr', 'TotRmsAbvGrd', '1stFlrSF', '2ndFlrSF', 'BsmtFinSF1',
    'YrSold', 'SaleCondition', 'GarageType', 'GarageYrBlt', 'Foundation',
    'BsmtExposure', 'CentralAir', 'HalfBath', 'LotFrontage', 'MasVnrArea','SalePrice'
]
```

```python
X_train_important = X_train[important_features]

X_test_important = X_test[important_features]
```

```python
X_train_important.head()
```

```python
X_train_important.isna().mean()
```

```python
X_train_important.dtypes
```

```python
cols_with_nans = X_train_important.columns[X_train_important.isna().any()].tolist()
cols_with_nans
```

# Droping every column with at least 1 nan value, just for easy preprocessing.

```python
X_train_important = X_train_important.drop(columns=cols_with_nans)
X_test_important = X_test_important.drop(columns=cols_with_nans)
```

```python
X_train_important
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_cols = X_train_important.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

```python
encoder = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
X_train_encoded = encoder.fit_transform(X_train_important)
```

```python
X_test_encoded = encoder.transform(X_test_important)
```

```python
encoded_feature_names = encoder.get_feature_names_out()
X_train = pd.DataFrame(X_train_encoded, columns=encoded_feature_names)
X_test = pd.DataFrame(X_test_encoded, columns=encoded_feature_names)
```

```python
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
```

```python
X_train
```

```python
X_test
```

# Training Part

```python

y_train = X_train.pop('remainder__SalePrice')
y_test = X_test.pop('remainder__SalePrice')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)

y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error

def evaluate(y_true, y_pred, dataset="Set"):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"{dataset} Performance:")
    print(f"MAE: {mae:.2f}")
    print(f"MAPE: {mape:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R² Score: {r2:.2f}\n")

# Evaluate Train and Test Set
evaluate(y_train, y_train_pred, "Train")
evaluate(y_test, y_test_pred, "Test")
```

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
rmse_scores = np.sqrt(-cross_val_score(model, X_train, y_train, cv=kf, 
                                     scoring='neg_mean_squared_error'))

print("RMSE for each fold:", rmse_scores)
print(f"Average RMSE: {np.mean(rmse_scores):.4f}")
```
